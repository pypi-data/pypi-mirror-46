<h3>InstrumentControl Examples</h3>
<ul><li>
<a href="./VNA_Data_Acquisition_Example.ipynb">Simple VNA Measurement Example</a><br/>
An example of taking scattering parameters and wave parameters from a VNA. 
</li>
<li>
<a href="./VNA_Measurement_Example_WR15.ipynb">WR15 VNA Measurement Example</a><br/>
An example of using the VNA class to measure one-port and two-port scattering and wave parameters and then save them. These measurements use the ZVA67 with a WR15 extension head at 50-75GHz.This is more of a lab example but shows the VNA class in action.  
</li>
<li>
<a href="./Emulation_Mode_Instrument_Control_Example.ipynb">Example of Emulated Instrument</a><br/>
An example of the EmulationInstrument mode that occurs when the device is not connected. It serves as a diagnostic to show the commands that get sent by any class. 
</li></ul>

<a id="Lab"></a>
<h2>Lab Examples </h2>
Examples pertinent to work in my laboratory.
<ul><li>
<a href="./Automating_MUF_Example.ipynb">Automating MUF Example</a><br/>
An example of running the MUF as a script in python. The example covers running several applications such as VNAUncert, Measurement and PostProcessor using pythonnet. Additionally, there is an example using pyMez to create a series of MUF results for a set of different Monte Carlo trials.
</li>
<li>
<a href="./Error_Correction_Comparison_Reciprocal_Non.ipynb">Comparison of using the StatistiCAL package and pyMez to apply corrections </a><br/>
A comparison of correcting from 12 term corrections and 8-term corrections both reciprocal and non-reciprocal generated by statistiCAL and applied by both pyMez and StatistiCAL. The ultimate result is they agree (statistiCAL only applies the reciprocal ones).  
</li>
<li>
<a href="./PNAGrabber_MUF_Example.ipynb">A work flow for PNA Grabber and Microwave Uncertainty Framework</a><br/>
An introduction to using the NIST* software in two-port microwave calibrations. 
</li>
<li>
<a href="./Data_Guidance_Group_01.ipynb">General Data Guidance for XML/Ascii Data </a><br/>
General data guidance for storing data more effectively in 672.01 
</li>
<li>
<a href="./Data_Transformation_Check_Standard.ipynb">Example of creating check standard and dut database </a><br/>
A detailed description of how to create a de-normalized database using the ascii.dut and converted check standard data.
</li>
<li>
<a href="./Current_Acquistion_Analysis_Routine_SParameter.ipynb">Description of our current data processing scheme in Sparameters </a><br/>
An outlined description of how we take data currently and places we need improvement. In addition, an outline of the imporvement needed at each step.
</li>
<li>
<a href="./Git_Work_Flow.ipynb">Git Work Flow For Single Contributor </a><br/>
An outline of my Git work flow for a single contributor.  
</li>
<li>
<a href="./Refactoring_VNAFreqV8_Example.ipynb">Refactoring a python program example </a><br/>
A set of thoughts and steps to refactor code to make it more shareable.
</li></ul>
